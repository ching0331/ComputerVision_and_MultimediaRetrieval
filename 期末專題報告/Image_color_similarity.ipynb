# 下載圖片

```python
import numpy as np
import os
import requests
from bs4 import BeautifulSoup

import random
import cv2
 
def download_images(url, start, end, folder, downloaded_images):
    web = requests.get(url) # 使用 requests取得網頁內容
    soup = BeautifulSoup(web.text, "html.parser") # 使用 beautifulsoup對取得的html內容進行解析
    for i in range(start, end):
        img_url = soup.select('div[class~="d95fI"] div[class~="WxXog"] img')[i]["src"] # 取得image_url
        # 可避免重複下載相同圖片
        #if img_url in downloaded_images:
            #continue
        # 下載圖片
        img_file = requests.get(img_url)
        with open(f'{folder}/{str(i+1).zfill(4)}.png', 'wb') as f:
            f.write(img_file.content)
        # 將圖片URL加入集合
        downloaded_images.add(img_url)
 
 
# 儲存已下載圖片的 URL 集合
downloaded_images = set()
 
os.makedirs('Unsplash/day_night', exist_ok=True)
download_images('https://unsplash.com/s/photos/day-sky?license=free', 0, 18, 'Unsplash/day_night', downloaded_images)
download_images('https://unsplash.com/s/photos/night?license=free', 18, 36, 'Unsplash/day_night', downloaded_images)
 
os.makedirs('Unsplash/sun_moon', exist_ok=True)
download_images('https://unsplash.com/s/photos/sun?license=free', 0, 18, 'Unsplash/sun_moon', downloaded_images)
download_images('https://unsplash.com/s/photos/moon?license=free', 18, 36, 'Unsplash/sun_moon', downloaded_images)
 
os.makedirs('Unsplash/ball', exist_ok=True)
download_images('https://unsplash.com/s/photos/tennis-ball?license=free', 0, 9, 'Unsplash/ball', downloaded_images)
download_images('https://unsplash.com/s/photos/basketball?license=free', 9, 18, 'Unsplash/ball', downloaded_images)
download_images('https://unsplash.com/s/photos/badminton-ball?license=free', 18, 27, 'Unsplash/ball', downloaded_images)
download_images('https://unsplash.com/s/photos/baseball?license=free', 27, 36, 'Unsplash/ball', downloaded_images)
 
os.makedirs('Unsplash/suit_dress', exist_ok=True)
download_images('https://unsplash.com/s/photos/Wedding-dress?license=free', 0, 18, 'Unsplash/suit_dress', downloaded_images)
download_images('https://unsplash.com/s/photos/suit?license=free', 18, 36, 'Unsplash/suit_dress', downloaded_images)
```

```python
# 將所有圖片放入串列
def get_image_list(folder_path):
    image_files = sorted([f for f in os.listdir(folder_path) if f.endswith(('.png', '.jpg', '.jpeg'))])
    images = []
    for file in image_files:
        img_path = os.path.join(folder_path, file)
        img = cv2.imread(img_path)
        if img is not None:
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            images.append(img_rgb)
    return images

gallery_images = get_image_list(image_list[num])

#--------------------------------------------------------------------

# 計算圖片的直方圖
def calculate_histograms(images):
    hists = []
    for img in images:
        hist = cv2.calcHist([img], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
        hist = hist.reshape(512)
        hists.append(hist)
    return hists

hists_image = calculate_histograms(gallery_images)

#--------------------------------------------------------------------

# 計算距離和排序

dist = np.zeros((36), np.float32)

for j in range(36):
    dist[j] = 0
    for k in range(512):
        dist[j] += abs(hist_q[k]-hists_image[j][k])
    print('dist[',j,'] =',dist[j])
        
sorted_indices = sorted(range(36), key=lambda k: dist[k])
print('sorted_indices = ',  sorted_indices)

# 顯示圖片
plt.imshow(query)
plt.xticks([]),plt.yticks([])
plt.show()

plt.figure(figsize=(20, 20))
for j in range(36):
    retrieved_index = sorted_indices[j]
    plt.subplot(6, 6, j+1)
    plt.imshow(gallery_images[retrieved_index])
    plt.xticks([]),plt.yticks([])
plt.show()

#--------------------------------------------------------------------

# 找到資料夾中的所有圖片路徑
folder_path = image_list[num]
gallery_images_paths = [os.path.join(folder_path, f) for f in sorted(os.listdir(folder_path)) if f.endswith(('.png', '.jpg', '.jpeg'))]

# 根據 sorted_indices 的順序重新排列圖片路徑
gallery_images_paths = [gallery_images_paths[index] for index in sorted_indices]
```

# 合成及下載圖片

```python
# 下載套件
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 讀取並調整 gallery images 大小
# gallery_images_paths為以排序過的總圖片
gallery_images = []
for img_path in gallery_images_paths:
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (query.shape[1], query.shape[0]))  # 調整大小
    gallery_images.append(img)

# 確認所有圖片大小一致
img_height, img_width, _ = query.shape
 
# 創建合成圖片的空白畫布（6*6）
combined_image = np.zeros((img_height * 6, img_width * 6, 3), dtype=np.uint8)
 
# 放置 query image
combined_image[:img_height, :img_width, :] = query
 
# 放置 gallery images
for i in range(6):
    for j in range(6):
        if i == 0 and j == 0:
            continue  # 已經放置 query image
        index = i * 6 + j - 1  # 因為 query image 佔用了一個位置
        combined_image[i*img_height:(i+1)*img_height, j*img_width:(j+1)*img_width, :] = gallery_images[index]

# 顯示合成圖片
plt.imshow(combined_image)
plt.xticks([]), plt.yticks([])
plt.show()
 
# 儲存合成圖片
cv2.imwrite("combined_image.jpg", cv2.cvtColor(combined_image, cv2.COLOR_RGB2BGR))
```

```python
import os
import cv2
import matplotlib.pyplot as plt

# 顯示同一資料夾中的所有圖片
def show_images_from_folder(folder_path):
    image_files = sorted([f for f in os.listdir(folder_path) if f.endswith(('.png', '.jpg', '.jpeg'))])

    images = []

    for file in image_files:
        img_path = os.path.join(folder_path, file)
        img = cv2.imread(img_path)
        if img is not None:
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            images.append(img_rgb)

    num_images = len(images)
    plt.figure(figsize=(20, 20))
    for i in range(num_images):
        plt.subplot(6, 6, i + 1)  # 假設最多36張圖片，6x6的網格
        plt.imshow(images[i])
        plt.xticks([]), plt.yticks([])
    plt.show()

# 使用自訂義函式來顯示 Unsplash/day_night 資料夾中的圖片
show_images_from_folder('Unsplash/day_night')
# show_images_from_folder('Unsplash/ball')
# show_images_from_folder('Unsplash/sun_moon')
# show_images_from_folder('Unsplash/suit_dress')
```
